```python
X_train.drop('Product_ID',axis=1,inplace=True)
X_test.drop('Product_ID',axis=1,inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
##Train your Model
from sklearn.linear_model import LinearRegression

lm = LinearRegression()
lm.fit(X_train, y_train)
print(lm.fit(X_train, y_train))
```

```python
LinearRegression(copy_X=True, fit_intercept=True, n_jobs=None,
         normalize=False)
```

```python
predictions = lm.predict(X_test)
print("Predicted purchases (in dollars) for new costumers:", predictions)
```

```python
predictions
```

```python
pd.DataFrame(predictions,columns=['Predictions'])
```

```python
from sklearn import metrics

print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, predictions)
print('R2 Score:', r2)
```

```python
from sklearn.ensemble import RandomForestRegressor
##from sklearn.model_selection import GridSearchCV
rf = RandomForestRegressor(n_estimators = 100, random_state = 0)

rf.fit(X_train, y_train)
predictions1 = rf.predict(X_test)
pd.DataFrame(predictions1,columns=['Predictions_1'])

from sklearn import metrics
from sklearn.metrics import r2_score

print('MAE:', metrics.mean_absolute_error(y_test, predictions1))
print('MSE:', metrics.mean_squared_error(y_test, predictions1))
r2_1 = r2_score(y_test, predictions1)
print('R2 Score:', r2_1)
```

```python
from sklearn.tree import DecisionTreeRegressor
tree = DecisionTreeRegressor().fit(X_train, y_train)

predictions2 = tree.predict(X_test)

print('MAE:', metrics.mean_absolute_error(y_test, predictions2))
print('MSE:', metrics.mean_squared_error(y_test, predictions2))
r2_2 = r2_score(y_test, predictions1)
print('R2 Score:', r2_2)
```
